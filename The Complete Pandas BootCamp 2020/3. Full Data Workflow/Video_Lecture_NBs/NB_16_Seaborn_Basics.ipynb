## Advanced / statistical plotting with Seaborn

### First Steps

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
# NUMBER OF OBSERVATIONS PER GENDER, X-Axis
plt.figure(figsize=(12,8))
sns.countplot(data = titanic, x = "sex")
plt.show()
```

```python
# NUMBER OF OBSERVATIONS PER GENDER, Y-Axis
plt.figure(figsize=(12,8))
sns.countplot(data = titanic, y = "sex")
plt.show()
```

```python
# NUMBER OF OBSERVATIONS PER GENDER, COLOUR BY CLASS (using a Palette colour)
plt.figure(figsize=(12,8))
sns.set(font_scale=2, palette= "viridis")
sns.countplot(data = titanic, x = "sex", hue = "pclass")
plt.show()
```

## Categorical Plots

```python
# STRIP PLOT: One variable must be categorical!
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
#sns.stripplot(data = titanic, x = "sex", y = "age", jitter = False, hue = None, dodge = False)
#sns.stripplot(data = titanic, x = "sex", y = "age", jitter = True, hue = None, dodge = False)
#sns.stripplot(data = titanic, x = "sex", y = "age", jitter = True, hue = "pclass", dodge = False)
sns.stripplot(data = titanic, x = "sex", y = "age", jitter = True, hue = "pclass", dodge = True)
plt.show()

# Jitter=True avoids the chart being one single line
# Dodge=True splits each hue into its own group
```

```python
# SWARM PLOT: Locates all points with the same 'y' value and aligns them
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.swarmplot(data = titanic, x = "sex", y = "age", hue = "pclass", dodge = True) # Warning due to NA's
plt.show()
```

```python
# VIOLIN PLOT: Shows frequency
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data = titanic, x = "sex", y = "age", hue = "pclass", dodge = True)
#sns.swarmplot(data = titanic, x = "sex", y = "age", hue = "pclass", dodge = True, color = "black")
plt.show()
```

```python
# VIOLIN + SWARM PLOTS: for increased visibility
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data = titanic, x = "sex", y = "age", hue = "pclass", dodge = True)
sns.swarmplot(data = titanic, x = "sex", y = "age", hue = "pclass", dodge = True, color = "black")
plt.show()
```

```python
# VIOLIN PLOT: Swapped variables -> Split=True merges plots in the same x-axis category
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data = titanic, x = "pclass", y = "age", hue = "sex", dodge = True, split = True )
plt.show()
```

```python
# BAR PLOT: Really shows 1st class passengers were older...
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.barplot(data = titanic, x = "pclass", y = "age", hue = "sex", dodge = True)
plt.show()
```

```python
# POINT PLOT: Shows the mean age for each x-axis category and the min-max variation
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.pointplot(data = titanic, x = "pclass", y = "age", hue = "sex", dodge = True) # Dodge=True to avoid overlapping!
plt.show()
```

## Jointplots / Regression

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

### Was age and fare valuable features to survive the Titanic?

```python
# JOINT PLOT: Point plot + Histogram
sns.set(font_scale=1.5)
sns.jointplot(data = titanic, x = "age", y = "fare", height = 8, kind = "reg") # kind is a regression for 95% of cases
#sns.jointplot(data = titanic, x = "age", y = "fare", height = 8, kind = "hex") # kind is a hex, i.e., shows quantities
#sns.jointplot(data = titanic, x = "age", y = "fare", height = 8, kind = "kde") # kind is a kernel density estimator
plt.show()
```

```python
# LM PLOT: Linear Regression Plot
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "fare", aspect= 1, height=8)
plt.show()
```

```python
# LM PLOT: Linear Regression Plot for each hue (Males and Females)
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "fare", aspect= 1, height=8, hue="sex")
plt.show()
```

```python
# LM PLOT: Linear Regression Plot (completely separate plots for each gender)
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "fare", aspect= 1, height=8, col = "sex")
plt.show()
```

```python
# LM PLOT: Correlation between "age" and "survivability", per gender
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "survived", aspect= 1, height=8, col = "sex", logistic= True)
plt.show()

# Because 'y' is a binary category, logistic=True!
```

```python
# LM PLOT: Correlation between "age" and "survivability", per class
sns.set(font_scale=1.5)
sns.lmplot(data = titanic, x = "age", y = "survived", aspect= 1, height=8, col = "pclass", logistic= True)
plt.show()
```

## Matrixplots / Heatmaps

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
pd.crosstab(titanic.sex, titanic.pclass) # Absolute Frequency
```

```python
# HEAT MAP: Visualize Crosstabs!
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
#sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass))
#sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass), annot= True, fmt = "d")
#sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass), annot= True, fmt = "d", cmap = "Reds")
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass), annot= True, fmt = "d", cmap = "Reds", vmax = 150)
plt.show()

#annot=True to show numbers
#fmt="d" to avoid scientific notation
#cmap="Reds" for color codes
#vmax for maximum value in the label, to balance the color map
```

```python
pd.crosstab(titanic.sex, titanic.pclass, values= titanic.survived, aggfunc= "mean")
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass, values= titanic.survived, aggfunc= "mean"), annot= True, cmap = "Reds")
plt.show()
```

```python
titanic.corr()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(titanic.corr(), annot= True, cmap = "Reds")
plt.show()
```
