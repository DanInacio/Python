### Advanced Aggregation with agg()

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv", usecols = ["survived", "pclass", "sex", "age", "fare"])
```

```python
titanic.head()
```

```python
titanic.groupby("sex").mean()
```

```python
titanic.groupby("sex").sum() # Only "survived" and "fare" make sense
```

```python
titanic.groupby("sex").agg(["mean", "sum", "min", "max"]) # 4 calculations in 1
```

```python
# Only relevant calculations per column!
titanic.groupby("sex").agg({"survived": ["sum", "mean"], "pclass": "mean", "age": ["mean", "median"], "fare": "max"})
```

### GroupBy Aggregation with Relabeling (new in Version 0.25)

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv", usecols = ["survived", "pclass", "sex", "age", "fare"])
```

```python
titanic.head()
```

```python
titanic.groupby("sex").survived.mean()
```

```python
# Python identifies if you want the name of a column or a mathematical function
titanic.groupby("sex").agg(survival_rate = ("survived", "mean"))
```

```python
# WARNING: Column names MUST NOT be named after mathematical functions!
titanic.groupby("sex").agg({"survived": ["sum", "mean"], "age": ["mean"]})
```

```python
titanic.groupby("sex").agg(survived_total = ("survived", "sum"), 
                           survival_rate = ("survived", "mean"),
                           mean_age = ("age", "mean"))
```

### Transformation with transform()

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
#titanic.groupby(["sex", "pclass"]).survived.mean() -> Returns mean survival rates per gender, per class

titanic.groupby(["sex", "pclass"]).survived.transform("mean")
```

The command transformed the "survived" column into the average chance of survival for each passenger, given gender and class!!!

```python
titanic["group_surv_rate"] = titanic.groupby(["sex", "pclass"]).survived.transform("mean") # New column
```

```python
titanic.head()
```

```python
titanic["outliers"] = abs(titanic.survived-titanic.group_surv_rate)
```

```python
titanic.head()
```

```python
titanic[titanic.outliers > 0.85]
```

### Generalizing split-apply-combine with apply()

#### How to get the 5 oldest people who survived, per gender?

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv", usecols = ["survived", "pclass", "sex", "age", "fare"])
```

```python
titanic.head()
```

```python
titanic.groupby("sex").mean()
```

```python
female_group = list(titanic.groupby("sex"))[0][1]
female_group
```

```python
female_group.mean().astype("float")
```

```python
def group_mean(group):
    return group.mean()
```

```python
group_mean(female_group)
```

```python
titanic.groupby("sex").apply(group_mean)
```

```python
titanic.nlargest(5, "age")
```

```python
def five_oldest_surv(group):
    return group[group.survived == 1].nlargest(5, "age")
```

```python
titanic.groupby("sex").apply(five_oldest_surv)
```

```python
# All who survived
#titanic.loc[titanic.survived == 1]

# All females/males who survived
#list(titanic.loc[titanic.survived == 1].groupby("sex"))[0][1]
#list(titanic.loc[titanic.survived == 1].groupby("sex"))[1][1]

# nlargest() DOES NOT EXIST IN DATAFRAMES!!!
titanic.loc[titanic.survived == 1].groupby("sex").nlargest(5,"age")
```

### stack() and unstack()

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
medals_by_country = summer.groupby(["Country", "Medal"]).Medal.count()
```

```python
medals_by_country
```

```python
#medals_by_country.loc["USA"]
medals_by_country.loc[("USA", "Gold")] # Multi-index locating
```

```python
medals_by_country.shape # 347 rows, 1 column
```

```python
medals_by_country.unstack(level = -1) # Last index is unstacked (-1), i.e., rows become columns
```

```python
medals_by_country = medals_by_country.unstack(level = -1, fill_value= 0)
```

```python
medals_by_country.head()
```

```python
medals_by_country.shape
```

```python
medals_by_country = medals_by_country[["Gold", "Silver", "Bronze"]] # Reordering columns
```

```python
medals_by_country.head()
```

```python
medals_by_country.sort_values(by = ["Gold", "Silver", "Bronze"], ascending = [False, False, False], inplace = True)
```

```python
medals_by_country.head(10)
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
medals_by_country.head(10).plot(kind = "bar", figsize = (12,8), fontsize = 13)
plt.xlabel("Country", fontsize = 13)
plt.ylabel("Medals", fontsize = 13)
plt.title("Medals per Country", fontsize = 16)
plt.legend(fontsize = 15)
plt.show()
```
